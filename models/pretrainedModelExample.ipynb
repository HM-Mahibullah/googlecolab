```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np

# Load ResNet50 model with pretrained ImageNet weights
model = ResNet50(weights='imagenet')

# Define target image path and load it with proper dimensions
# ERROR: FileNotFoundError. Please upload the image file 'Original-25901-aw609563.jpg'
# to your Colab environment or change 'img_path' to an existing image file.
img_path = '/content/dog.png' # <--- This file was not found
img = image.load_img(img_path, target_size=(224, 224))

# Convert image to numpy array, expand dimensions, and preprocess
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Perform prediction and decode top 3 results
preds = model.predict(x)
print('Predicted:', decode_predictions(preds, top=3)[0])
```
